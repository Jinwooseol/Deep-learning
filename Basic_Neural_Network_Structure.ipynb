#Fully-Connected Layer ÏàòÌïôÏ†Å ÌëúÌòÑ
- Î™®Îì† NodesÍ∞Ä ÏÑúÎ°ú edgeÎ°ú Ïó∞Í≤∞Îêú ÏÉÅÌÉú

- Weight set $W = [w_0, w_1, ..., w_{M-1}]^T$ ÌòïÌÉúÎ°ú  
  Í∞ÅÍ∞ÅÏùò $w_k$Îäî $N$ x 1 ÌòïÌÉúÏùò vector, ($W$Îäî $N$ x $M$ Matrix)
  
- Bias $b = [b_0, b_1, ..., b_{M-1}]$

- $y_0 = f(w^T_0x + b_0)$

- $y_1 = f(w^T_1x + b_1)$

- $y_2 = f(w^T_2x + b_2)$

  ...

- $y_{M-1} = f(w^T_{M-1}x + b_{M-1})$

‚Üí $y = f(Wx + b)$

#AND gate
- AND gateÎ•º ÎßåÏ°±ÏãúÌÇ§Îäî weights, bias Íµ¨ÌïòÍ∏∞

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

```python
def AND(a, b):
  input = np.array([a, b])
  weight = np.array([0.4, 0.4])
  bias = -0.6
  value = np.sum(input * weight) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = -0.6

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

plt.plot(x1, y, 'r--')  # r--ÏùÄ red Ï†êÏÑ†, and gateÎ•º Í≤∞Ï†ïÌïòÍ∏∞ ÏúÑÌïú Í≤ΩÍ≥ÑÏÑ†
plt.scatter(0, 0, color='black', marker='x', s=150)  # s is size
plt.scatter(0, 1, color='black', marker='x', s=150)
plt.scatter(1, 0, color='black', marker='x', s=150)
plt.scatter(1, 1, color='orange', marker='o', s=150)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.show()
```

#OR Gate
- OR gateÎ•º ÎßåÏ°±ÏãúÌÇ§Îäî weightsÏôÄ bias Íµ¨ÌïòÍ∏∞

```python
def OR(a, b):
  input = np.array([a, b])
  weight = np.array([0.4, 0.5])
  bias = -0.3
  value = np.sum(input * weight) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

# NAND Gate

```python
def NAND(a, b):
  input = np.array([a, b])
  weight = np.array([-0.6, -0.5])
  bias = 0.7
  value = np.sum(input * weight) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

#Multi Layer Perceptron
- Input layerÍ≥º output layer ÏÇ¨Ïù¥Ïóê ÌïòÎÇòÏù¥ÏÉÅÏùò hidden layerÍ∞Ä Ï°¥Ïû¨  
  (5Í∞ú Ïù¥ÏÉÅÏùò hidden layerÍ∞Ä Ï°¥Ïû¨ÌïòÎ©¥ Deep Neural Network)

  
>Input layer ‚Üí hidden layer
- $z = f_Z(W_Lx + b_L)$

>Hidden layer ‚Üí output layer  
- $y = a_K(W_Kz + b_K)$

#XOR Gate ÌëúÌòÑ

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```python
print(XOR(0, 0))
print(XOR(0, 1))
print(XOR(1, 0))
print(XOR(1, 1))
```

#Step Function
$y = \left\{\begin{matrix}0
 & (x < 0) \\1
 &  (x \geq 0)\\
\end{matrix}\right.$

```python
def step_function(x):
  if x > 0:
    return 1
  else:
    return 0
```

```python
def step_function_numpy(x):
  y = x > 0
  return y.astype(np.int)
```

```python
print(step_function(-3))
print(step_function(5))

a = np.array([5, 3, -4, 2.0])
print(step_function_numpy(a))
```

#Sigmoid Function
- Binary classificationÏóêÏÑú Ï£ºÎ°ú ÏÇ¨Ïö©  
  * ÎßàÏßÄÎßâ Ï∂úÎ†•Ï∏µÏùò activation functionÏúºÎ°ú ÏÇ¨Ïö©
- Ï∂úÎ†•Í∞íÏùÄ 0 ~ 1ÏÇ¨Ïù¥, ÌôïÎ•†Î°ú ÌëúÌòÑ Í∞ÄÎä•

$y = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
print(sigmoid(3))
print(sigmoid(-3))
```

```python
x = np.arange(-10, 10, 0.01)
y = sigmoid(x)

plt.plot(x, y, 'r-')
plt.show()
```

#ReLU(Rectified Linear Unit)
- Í∞ÄÏû• ÎßéÏù¥ Ïì∞Ïù¥Îäî functionÏ§ë ÌïòÎÇò
- 0 Ïù¥Ï†ÑÍπåÏßÄÎäî 0ÏúºÎ°ú Í∞í Ïú†ÏßÄ, 0Î≥¥Îã§ ÌÅ¨Î©¥ input value
- $y = \left\{\begin{matrix}0
 & (x \le 0) \\x
 &  (x > 0)\\
\end{matrix}\right.$

```python
def ReLU(x):
  if x > 0:
    return x
  else:
    return 0
```

```python
def ReLU_numpy(x):
  return x * (x > 0)
```

```python
print(ReLU(5))
print(ReLU(-3))
```

```python
x = np.arange(-5.0, 5.0, 0.01)
y = ReLU_numpy(x)

plt.plot(x, y, 'r-')
plt.show()
```

#tanh (Hyperbolic tangent function)
- $y = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
print(tanh(3))
print(tanh(-3))
```

```python
x = np.arange(-5, 5, 0.01)
y = tanh(x)

plt.plot(x, y, 'r-')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

#Softmax
- Multi class classificationÏóêÏÑú ÏÇ¨Ïö©
- Input valueÏùò ÏòÅÌñ•ÏùÑ ÌÅ¨Í≤å Î∞õÏùå
- Output valueÎ•º **ÌôïÎ•†**Ïóê ÎåÄÏùëÍ∞ÄÎä•
- Output valueÏùò Ï¥ùÌï©ÏùÄ 1
- $y_k = \frac{e^{a_k}}{Œ£_je^{a_j}}$

```python
def softmax(a):
 exp_a = np.exp(a)
 sum_exp_a = np.sum(exp_a)
 y = exp_a / sum_exp_a
 return y 
```

```python
a = np.array([0.3, 0.2, 4.0, -1.2])
print(softmax(a))
print(np.sum(softmax(a)))
```

#SoftmaxÏùò Ï£ºÏùòÏ†ê
- Overflow Î¨∏Ï†ú Î∞úÏÉù
- Exponential functionÏùÑ ÏÇ¨Ïö©ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê input valueÍ∞Ä ÎÑàÎ¨¥ ÌÅ∞ Í≤ΩÏö∞, Î¨¥ÌïúÎåÄ (inf)Í∞Ä Î∞òÌôò

  Í∞úÏÑ†Îêú ÏàòÏãù
  - $y_k = \frac{Ce^{a_k}}{CŒ£_je^{a_j}}$ = $\frac{e^{(a_k + logC)}}{Œ£_je^{(a_j + logC)}}$ = $\frac{e^{(a_k + C')}}{Œ£_je^{(a_j + C')}}$
  (C'Îäî -(Í∞ÄÏû• ÌÅ∞ ÏûÖÎ†•Í∞í))

```python
a = np.array([500, 600, 700, 1000])
print(softmax(a))  # Overflow Î∞úÏÉù
```

```python
def adv_softmax(a):
  C = np.max(a)
  return(np.exp(a - C) / np.sum(np.exp(a - C)))
```

```python
a = np.array([500, 600, 700, 1000])
print(adv_softmax(a))
```

```python
x = np.arange(-5.0, 5.0, 0.01)
y = adv_softmax(x)

plt.plot(x, y, 'r-')
plt.show()
```

#LeakyReLU
- ReLUÏùò Î≥ÄÌòïÏúºÎ°ú 0Ïù¥ÌïòÏùò ÏûÖÎ†•Ïóê ÎåÄÌï¥ÏÑúÎèÑ Í∞íÏùò Î≥ÄÌôî Ï°¥Ïû¨
- $f_a(x) = \left\{\begin{matrix}x
 & (x \ge 0) \\ax
 &  (x < 0)\\
\end{matrix}\right.$

```python
def LeakyReLU(x):
  a = 0.1
  return np.maximum(a*x, x)
```

```python
x = np.arange(-5.0, 5.0, 0.01)
y = LeakyReLU(x)

plt.plot(x, y, 'r-')
plt.show()
```

#ELU (Exponential Linear Units)
- $f(Œ±, x) = \left\{\begin{matrix}x
 & (x > 0) \\Œ±(e^x - 1)
 &  (x ‚â§ 0)\\
\end{matrix}\right.$

```python
def ELU(x):
  alpha = 1.0
  return (x >= 0) * x + (x < 0) * alpha * (np.exp(x) - 1)
```

```python
x = np.arange(-5.0, 5.0, 0.01)
y = ELU(x)

plt.plot(x, y, 'r-')
plt.axhline(y=0, color='k')
plt.show()
```

# Three Layers Neural Network (practice)
- 2 class Î∂ÑÎ•ò

- **Input Layer**  
  - Number of Nodes: 3

  * First hidden layer

#Define Activation Function

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
# Input layer -> first hidden layer
X = np.array([1.0, 0.5, 0.4])
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6], [0.3, 0.5, 0.7]])
B1 = np.array([1, 1, 1])

print("X's shape: ", X.shape)
print("W1's shape: ", W1.shape)
print("B1's shape: ", B1.shape)

A1 = np.dot(X, W1) + B1   # activation function ÌÜµÍ≥º Ï†Ñ
Z1 = sigmoid(A1)          # activation function ÌÜµÍ≥º ÌõÑ

print(A1)
print(Z1)
```

```python
# First hidden layer -> second hidden layer
W2 = np.array([[0.2, 0.4, 0.6], [0.1, 0.3, 0.5], [0.4, 0.6, 0.8]])
B2 = np.array([1, 1, 1])

print("W2's shape: ", W2.shape)
print("B2's shape: ", B2.shape)

A2 = np.dot(Z1, W2) + B2
Z2 = sigmoid(A2)

print(A2)
print(Z2)
```

```python
# Second hidden layer -> Output layer
W3 = np.array([[0.1, 0.3], [-0.1, -0.5], [0.3, 0.5]])
B3 = np.array([1, 1])

print("W3's shape: ", W3.shape)
print("B3's shape: ", B3.shape)

A3 = np.dot(Z2, W3) + B3
Z3 = sigmoid(A3)

print(A3)
print(Z3)
```

```python
# Output layer -> Output
W4 = np.array([[0.1, 0.2], [0.3, 0.5]])
B4 = np.array([1, 1])

print("W4's shape: ", W4.shape)
print("B4's shape: ", B4.shape)

A4 = np.dot(Z3, W4) + B4
Z4 = sigmoid(A4)

print(A4)
print(Z4)
```

#ÌïòÎÇòÎ°ú Ìï©ÏπòÍ∏∞

```python
def network():
  
  network = {}
  
  # First Layer 
  network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6], [0.3, 0.5, 0.7]])
  network['B1'] = np.array([1, 1, 1])

  # Second Layer 
  network['W2'] = np.array([[0.2, 0.4, 0.6], [0.1, 0.3, 0.5], [0.4, 0.6, 0.8]])
  network['B2'] = np.array([1, 1, 1])

  # Third Layer 
  network['W3'] = np.array([[0.1, 0.3], [-0.1, -0.5], [0.3, 0.5]])
  network['B3'] = np.array([1, 1])

  # Forth Layer 
  network['W4'] = np.array([[0.1, 0.2], [0.3, 0.5]])
  network['B4'] = np.array([1, 1])

  return network
```

```python
def forward(network, x):
  W1, W2, W3, W4 = network['W1'], network['W2'], network['W3'], network['W4']
  B1, B2, B3, B4 = network['B1'], network['B2'], network['B3'], network['B4']

  # (Input layer -> first hidden layer)
  A1 = np.dot(x, W1) + B1
  Z1 = sigmoid(A1)

  # (First hidden layer -> second hidden layer)
  A2 = np.dot(Z1, W2) + B2
  Z2 = sigmoid(A2)

  # (Second hidden layer -> Output layer)
  A3 = np.dot(Z2, W3) + B3
  Z3 = sigmoid(A3)

  # (Output layer -> Output)
  A4 = np.dot(Z3, W4) + B4
  y = sigmoid(A4)

  return y
```

#Neural Network Ï∂îÎ°† Ïã§Ìñâ

```python
net = network()
x = np.array([0.3, 1.3, -2.2])
y = forward(net, x)
print(y)
```
